### Initial compressor and turbine design:

```python
import pandas as pd 
from Project2BackendCalcs import TurboMachineryComputation
import IPython.display as disp
import numpy as np
from IPython.display import Markdown as md

##### Compressor initial design #####
backend = TurboMachineryComputation()
N,rt,rm,rr,M1t,rt3,rr3, P_02, T_02, P_03, T_03 = backend.update()

print('Rotational speed: '+str(np.round(N,2))+' (rev/s)')
print('Tip radius (inlet): '+str(np.round(rt,5))+' (m)')
print('Mean radius (constant): '+str(np.round(rm,5))+' (m)')
print('Root radius (inlet): '+str(np.round(rr,5))+' (m)')
print('Tip radius (outlet): '+str(np.round(rt3,5))+' (m)')
print('Root radius (outlet): '+str(np.round(rr3,5))+' (m)')
# print('Compressor Outlet Pressure: ' +str(np.round(P_02,5)))
# print('Compressor Outlet Temp: ' +str(np.round(T_02,5)))
# print('TIP: ' +str(np.round(P_03,5)))
# print('TIT: ' +str(np.round(T_03,5)))
print('Mach number at tip (inlet): '+str(np.round(M1t,3)))

meantable,sizingtable,tiproot_table,whirl_table,vel_table,dif_table = backend.fullcompressor()

# disp.display(table)
print('\nMean radius values for the compressor are: ')
disp.display(meantable)

print('\nSizing throughout the compressor: ')
disp.display(sizingtable)

print('\nTip and root angles: ')
disp.display(tiproot_table)

print('\nWhirl velocities: ')
disp.display(whirl_table)

print('\nAbsolute and relative velocities: ')
disp.display(vel_table)

print('\nDiffusion (de Haller criteria): ')
disp.display(dif_table)

##### Turbine initial design #####
gasParamsDF, measurementsDF, rootDF, tipDF, rtMeasurements, stage_est, Um = backend.fullturbine()
print('\nMean Blade Speed: {} m/s '.format(Um))
print('Tip radius (inlet): '+str(rtMeasurements[1])+' (m)')
print('Mean radius (constant): '+str(rtMeasurements[2])+' (m)')
print('Root radius (inlet): '+str(rtMeasurements[0])+' (m)')
print('Tip radius (outlet): '+str(rtMeasurements[4])+' (m)')
print('Root radius (outlet): '+str(rtMeasurements[3])+' (m)')

# print('\nEstimated No. of Stages: ', stage_est )
print('\nGas Parameters for the Turbine: ')
disp.display(gasParamsDF)

print('\nSizing through the turbine: ')
disp.display(measurementsDF)

print('\nRoot Parameters for the Turbine: ')
disp.display(rootDF)

print('\nTip Parameters for the turbine: ')
disp.display(tipDF)
```
